```python
#ambil data MCA & DennySiregar dari 25 Nov hingga 28 Nov
mca_twits = []
den_twits = []
start_date = datetime(2020, 11, 25, 0, 0, 0)
end_date = datetime(2020, 11, 28, 0, 0, 0)

for item in list_MCAOps:
    c_date = parse(item['created_at'])
    c_date = c_date.strftime('%Y-%m-%d %H:%M:%S')
    c_date = datetime.strptime(c_date,'%Y-%m-%d %H:%M:%S')
    if c_date >= start_date and c_date <= end_date:
        id_str = item['id_str']
        full_text = item['full_text']
        created_at = item['created_at']
        favorite_count = item['favorite_count']
        retweet_count = item['retweet_count']
        is_retweeted = False
        retweeted_from = ""
        try:
            retweeted_from = item['retweeted_status']['user']['screen_name']
            is_retweeted = True
            look_up_id = [item['retweeted_status']['id_str']]
            look_up_twit = api.statuses_lookup(look_up_id)[0]._json
            favorite_count = look_up_twit['favorite_count']
        except:
            pass
        temp_list = [id_str, full_text, created_at, favorite_count, retweet_count, is_retweeted, retweeted_from]
        mca_twits.append(temp_list)

for item in list_DennySiregar7:
    c_date = parse(item['created_at'])
    c_date = c_date.strftime('%Y-%m-%d %H:%M:%S')
    c_date = datetime.strptime(c_date,'%Y-%m-%d %H:%M:%S')
    if c_date >= start_date and c_date <= end_date:
        id_str = item['id_str']
        full_text = item['full_text']
        created_at = item['created_at']
        favorite_count = item['favorite_count']
        retweet_count = item['retweet_count']
        is_retweeted = False
        retweeted_from = ""
        try:
            retweeted_from = item['retweeted_status']['user']['screen_name']
            is_retweeted = True
            look_up_id = [item['retweeted_status']['id_str']]
            look_up_twit = api.statuses_lookup(look_up_id)[0]._json
            favorite_count = look_up_twit['favorite_count']
        except:
            pass
        temp_list = [id_str, full_text, created_at, favorite_count, retweet_count, is_retweeted, retweeted_from]
        den_twits.append(temp_list)

mca_df = pd.DataFrame(mca_twits, columns=['id_str','full_text','created_at','favorite_count','retweet_count','is_retweeted','retweeted_from'])
den_df = pd.DataFrame(den_twits, columns=['id_str','full_text','created_at','favorite_count','retweet_count','is_retweeted','retweeted_from'])
```

```python
mca_df.to_csv('mca.csv',index=False)
den_df.to_csv('den.csv',index=False)
```
